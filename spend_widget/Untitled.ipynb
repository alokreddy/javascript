```python
import json
from pandas.io.json import json_normalize
import pandas as pd
```

```python
with open('spend.json') as data_file:    
    data = json.load(data_file)  

df = json_normalize(data=data['data'], record_path = ['spendbydaterange'])
```

```python
df.columns
```

```python
df = df.apply(pd.to_numeric, errors='ignore')
```

```python
df.groupby(['spend_date'])['regular_billrate', 'overtime_billrate', 'total_billrate'].sum()
```
